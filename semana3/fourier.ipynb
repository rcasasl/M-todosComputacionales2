```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-6,6,1000)
L_value = 1


def fourier_sum(n, L, x):
    f_sum = np.zeros((len(x)))
    for i in range(1, n+1):
        f_sum += (2/(i*np.pi))*(1 - (-1)**i)*np.sin(i*np.pi*x/L)
    return f_sum  
y = fourier_sum(100, L_value, x)

plt.plot(x,y)
        
```
